```python
import glob
onlyfiles = glob.glob('C:/Users/ankur/Desktop/idbm/New Delhi Front Page/*-01.jpg')
def img_to_txt(x):
    s = x.split('/')
    s = s[4].split("-")[0]
    im = cv2.imread (x)
    im = cv2.bilateralFilter(np.array(im,dtype=np.uint8),9,75,75)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im = cv2.adaptiveThreshold(im, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 2)
    im = Image.fromarray(im)
    cropped_im_sl = im.crop(crop_rectangle_sl)
    c_sl = image_to_string(cropped_im_sl)
    cropped_im_pn = im.crop(crop_rectangle_pn)
    c_pn = image_to_string(cropped_im_pn)
    cropped_im_mt = im.crop(crop_rectangle_mt)
    c_mt = image_to_string(cropped_im_mt)
    cropped_im_wn = im.crop(crop_rectangle_wn)
    c_wn = image_to_string(cropped_im_wn)
    cropped_im_ps = im.crop(crop_rectangle_ps)
    c_ps = image_to_string(cropped_im_ps)
    cropped_im_teh = im.crop(crop_rectangle_teh)
    c_teh = image_to_string(cropped_im_teh)
    cropped_im_dist = im.crop(crop_rectangle_dist)
    c_dist = image_to_string(cropped_im_dist)
    cropped_im_pin = im.crop(crop_rectangle_pin)
    c_pin = image_to_string(cropped_im_pin)
    cropped_im_psn = im.crop(crop_rectangle_psn)
    c_psn = image_to_string(cropped_im_psn)
    cropped_im_psa = im.crop(crop_rectangle_psa)
    c_psa = image_to_string(cropped_im_psa)
    cropped_im_male = im.crop(crop_rectangle_male)
    c_male = image_to_string(cropped_im_male)
    cropped_im_fem = im.crop(crop_rectangle_fem)
    c_fem = image_to_string(cropped_im_fem)
    cropped_im_tg = im.crop(crop_rectangle_tg)
    c_tg = image_to_string(cropped_im_tg)
    cropped_im_tot = im.crop(crop_rectangle_tot)
    c_tot = image_to_string(cropped_im_tot)
    return c_sl,c_pn,c_mt,c_wn, c_ps,c_teh, c_dist, c_pin,c_psn,c_psa,c_male,c_fem,c_tg, c_tot,s
```

```python
state_legis = []
part_no = []
main_town = []
ward_no =[]
police_station =[]
tehsil = []
district = []
pin = []
PS_name = []
PS_add = []
male = []
female = []
third_g = []
total = []
file_name = []
for i in onlyfiles:
    a,b,c,d,e,f,g,h,t,j,k,l,m,n,s = img_to_txt(i)
    state_legis.append(a)
    part_no.append(b)
    main_town.append(c)
    ward_no.append(d)
    police_station.append(e)
    tehsil.append(f)
    district.append(g)
    pin.append(h)
    PS_name.append(t)
    PS_add.append(j)
    male.append(k)
    female.append(l)
    third_g.append(m)
    total.append(n)
    file_name.append(s)
```

```python
import numpy as np
front_page = pd.DataFrame(np.column_stack([state_legis, part_no, main_town,ward_no,police_station,tehsil,district,
                                           pin,PS_name,PS_add,male,female,third_g, total,file_name]), 
                               columns=['State Legislature', 'Part Number', 'Main Town', 'Ward Num', 'Police Station', 'Tehsil',
                                        'District', 'Pin','PS_name','PS_Add','Male','Female', 'Third Gender','Total','File_Name'])
```
